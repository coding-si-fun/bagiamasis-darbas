#### Apsivalome data nuo NaN, bei briskeriame žinučių ilgius.

```python
# nusikopijuojame 'teksto_analiziai_su_kalba.csv' į šitą direktoriją iš bendros. 
import pandas as pd

raw_df=pd.read_csv("data_for_chatBot.csv")

# prisdedame zinučių ilgius
raw_df["Žinučių_ilgiai"] = raw_df["Message"].str.len()

# išrušiuojame didėjnačia tvarka
# raw_df.sort_values('Žinučių_ilgiai', ascending=False)

# Patikriname ar yra tuščių reikšmių
raw_df.isna().sum() # 3463

# Numetame tas eilutes
numesta_NaN=raw_df.dropna()

# Numetame indeksą "Unnamed:0"
numesta_NaN.drop(["Unnamed: 0"], axis=1, inplace=True)
numesta_NaN.reset_index()
```

#### Numetame nereikalingus duomenis, tokius kaip vieno simbolio žinutes.

```python
# # Pasidaudojus analize pastebbėjau, kad yra daug nereikalingų, sistemos sukurtų žinučių...

sistemos_zinutes=numesta_NaN.loc[numesta_NaN['Message'].str.contains('Vaizdo skambutis baigėsi.|praleidote|praleido|atsiuntė priedą.|Jūs skambinote|Jums skambino|Dabar esate prisijungę prie „Messenger"|Dabar galite siųsti pranešimus ir paskambinti|pridėjo jus programoje „Messenger|Praleistas skambutis nuo|https|http',case=False, na=False)].index
numestos_sistemos_zinutes = numesta_NaN.drop(sistemos_zinutes); 
```

```python
# Pasikeičiame "default" nustatymus, kad galėtumėme matyti, daugiau eiličių.
pd.set_option('display.max_rows', 500)

numestos_sistemos_zinutes
```

```python

# Atrenku, žinučių skaičių, nuo "Sender" daugiai 2, taip išvalnat žmones, kurie rašė man kartą ir jiems neatsakiau. arba atskiau, bet tolimesnio kantokto nebuvo.


zinutes_skaiciuojamos = numestos_sistemos_zinutes.groupby('Sender')['Message'].count().reset_index()

zinutes_skaiciuojamos.columns = ['Sender', 'žinučių_kiekis']

siuntejai_su_zinutemis =zinutes_skaiciuojamos[zinutes_skaiciuojamos['žinučių_kiekis'] >= 3]['Sender']

isfiltruota = numestos_sistemos_zinutes[numestos_sistemos_zinutes['Sender'].isin(siuntejai_su_zinutemis )]

isfiltruota
```

```python
df=isfiltruota
# Convert 'TimeStamp' to datetime format
df['TimeStamp'] = pd.to_datetime(df['TimeStamp'])

# Calculate the time difference in seconds and take the absolute value
df['Skirtumas'] = df['TimeStamp'].diff().dt.total_seconds().abs()

# Convert the 'Skirtumas' column to a standard format (non-scientific)
df['Skirtumas'] = df['Skirtumas'].apply(lambda x: f"{x:.2f}" if pd.notnull(x) else x)

# Display the DataFrame with Skirtumas
print("DataFrame with Skirtumas (in seconds):")
df



```

```python
# Convert 'TimeStamp' to datetime
df['TimeStamp'] = pd.to_datetime(df['TimeStamp'])

# Initialize list to collect grouped messages
grouped_data = []

# Temporary variables to hold current group data
current_sender = df.loc[0, 'Sender']
current_message = df.loc[0, 'Message']
current_length = df.loc[0, 'Žinučių_ilgiai']
current_time = df.loc[0, 'TimeStamp']

# Iterate through DataFrame rows
for i in range(1, len(df)):
    row = df.iloc[i]
    
    # Check if sender is the same and time difference is within threshold (120 seconds in this case)
    if row['Sender'] == current_sender:
        # Combine messages and add length
        current_message += ", " + row['Message']
        current_length += row['Žinučių_ilgiai']
    else:
        # Save the current group
        grouped_data.append({
            "Sender": current_sender,
            "TimeStamp": current_time,
            "Message": current_message,
            "Žinučių_ilgiai": current_length
        })
        
        # Reset for the new group
        current_sender = row['Sender']
        current_message = row['Message']
        current_length = row['Žinučių_ilgiai']
        current_time = row['TimeStamp']

# Add the final group
grouped_data.append({
    "Sender": current_sender,
    "TimeStamp": current_time,
    "Message": current_message,
    "Žinučių_ilgiai": current_length
})

# Convert to DataFrame
grouped_df = pd.DataFrame(grouped_data)
# grouped_df.head(300)
# grouped_df.drop(0).iloc[60:267].iloc[::-1].reset_index(drop=True).to_csv("testui_eiliciu_botui.csv")
lentele=grouped_df.iloc[::-1].reset_index(drop=True)#.to_csv("visas_sugrupuotas_botui_apmokyti.csv")
# apversta=grouped_df.iloc[60:267].iloc[::-1].reset_index(drop=True);apversta
lentele.to_csv("visas_sugrupuotas_botui_apmokyti.csv")
lentele
```

```python
data=lentele

inputs = []
outputs = []

# Loop through each message in the dataset
for i in range(len(data) - 1):
    current_sender = data['Sender'][i]
    next_sender = data['Sender'][i + 1]
    
    # If the current sender is not Rimantas and the next sender is Rimantas, create an Input-Output pair
    if current_sender != "Rimantas Preiksas" and next_sender == "Rimantas Preiksas":
        inputs.append(data['Message'][i])
        outputs.append(data['Message'][i + 1])

# Create a new DataFrame with the structured Input and Output columns
structured_data = pd.DataFrame({
    'Input': inputs,
    'Output': outputs
})

# Display the first few rows of the new DataFrame


# Save the new DataFrame to a CSV file if needed
structured_data.to_csv('structured_conversation_data.csv', index=False)
structured_data.head(200)
```
